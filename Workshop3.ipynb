### 3. Joins

Joins are very useful for data cleaning. If the desired data is scattered around different dataframes, you would use the joins to merge dataframes and get a desired dataframe you want.

When you do a join, you need "join key(s)". What is a join key?

Join key(s) is column(s) containing common values. You would use this join key(s) to combine dataframes into one. Joins are useful when there is a "lookup table" with additional data you want to include on the other.

There are three big types of joins: inner join, left join and right join.

If you are familiar with SQL database, the joins are similar to SQL. If you have worked with excel spreadsheets, you can think of it as similar to vlookup.

In the example below, `species_df` is the look up table. The dataframe contains the genus, species and taxa for 55 species. `species_id` code is unique for each row. `survey_df` has a `species_id` too, and when `survey_df` is joined with `species_df` based on `species_id` (key), then I would have the genus, species and taxa information.

```python
survey_df.head() # Check survey_df
```

```python
species_df.head() # Check species_df
```

#### Inner join

Inner join combines two dataframes based on join key(s), and returns a new dataframe that contain __*only*__ those rows that have matching values on __*both*__ of the original dataframes.

```python
subset_species = species_df.head() # Sliced the species_df and called it "subset_species"
```

```python
# Performed an inner join
# subset_species has 5 species_id values
# inner join is the default for pd.merge
# specify first dataframe (left), second dataframe (right), and specify the join key column name for each dataframe
merge_inner = pd.merge(survey_df, subset_species, left_on = 'species_id', right_on = 'species_id')
```

```python
merge_inner # inner join result
```

What we can see is on result `merge_inner` dataframe is a set of rows that have the same `species_id` value on _both_ dataframes (`survey_df` and `subset_species`). 

`subset_species` has five `species_id` values (AB, AH, AS, BA, CB). Because two dataframes are _inner_ joined, only the rows that have the same five `species_id` values (AB, AH, AS, BA, CB) on `survey_df` dataframe will be in the joined result dataframe. 

#### Left join

Left join combines two dataframes based on join key(s) like an inner join. But, left join will return __*all*__ the rows from the _left_ dataframe, even those rows that do not have matching values in the right dataframe. If there is no match on the right dataframe, result will have a null value. 

I will subset the `survey_df` dataframe to illustrate as `surveys_subset` and illustrate left join.

```python
surveys_subset = survey_df.head(30) # slice the top 30 rows
```

```python
surveys_subset #check
```

```python
# Do a left join. specify it as left join by how = 'left' in the function.
merge_left = pd.merge(left = surveys_subset, right = subset_species, how = 'left', left_on = 'species_id', right_on = 'species_id')
```

```python
merge_left
```

The right dataframe to merge, `subset_species` has five species_id, AB, AH, AS, BA, and CB. The left dataframe to merge, `surveys_subset` does not have spcies_id that are AB, AH, AS, BA and CB. That is why null values are returned for columns genus, species and taxa.

```python
# This is another left join
# This time, dataframe on the right is species_df. species_df has *all* the species_id values 
merge_left = pd.merge(left = surveys_subset, right = species_df, how = 'left', left_on = 'species_id', right_on = 'species_id')
```

```python
merge_left # result
```

As you can see on the result, since `species_df` has all `species_id` values, so there are no null value for columns genus, species and taxa.
